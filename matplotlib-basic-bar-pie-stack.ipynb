```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.style.use('default')
plt.plot(ages, dev_salaries, color = "#444444", linestyle="--", label = 'All Devs')
plt.plot(ages, py_salaries, label = 'Python')
plt.tight_layout()
plt.legend()
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')
```

```python
plt.figure(figsize=(5,3))
plt.style.use('default')
plt.plot(ages, dev_salaries, color = "#444444", linestyle="--", label = 'All Devs')
plt.plot(ages, py_salaries, label = 'Python')
plt.fill_between(ages, py_salaries)
plt.tight_layout()
plt.legend()
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')
```

```python
plt.figure(figsize=(5,3))
plt.plot(ages, dev_salaries, color = "#444444", linestyle="--", label = 'All Devs')
plt.plot(ages, py_salaries, label = 'Python')
plt.fill_between(ages, py_salaries, alpha=0.25)
plt.tight_layout()
plt.legend()
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')
```

```python
plt.figure(figsize=(5,3))
plt.plot(ages, dev_salaries, color = "#444444", linestyle="--", label = 'All Devs')
plt.plot(ages, py_salaries, label = 'Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, alpha=0.25)
plt.tight_layout()
plt.legend()
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')
```

```python
plt.figure(figsize=(5,3))
plt.plot(ages, dev_salaries, color = "#444444", linestyle="--", label = 'All Devs')
plt.plot(ages, py_salaries, label = 'Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries > overall_median), 
                 interpolate=True, alpha=0.25)
plt.tight_layout()
plt.legend()
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')
```

```python
plt.figure(figsize=(5,3))
plt.plot(ages, dev_salaries, color = "#444444", linestyle="--", label = 'All Devs')
plt.plot(ages, py_salaries, label = 'Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries > overall_median), 
                 interpolate=True, alpha=0.25)
plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries <= overall_median), 
                 interpolate=True, color='red', alpha=0.25)
plt.tight_layout()
plt.legend()
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')
```

```python
plt.figure(figsize=(5,3))
plt.plot(ages, dev_salaries, color = "#444444", linestyle="--", label = 'All Devs')
plt.plot(ages, py_salaries, label = 'Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries > dev_salaries), label='Above Avg', 
                 interpolate=True, alpha=0.25)
plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries <= dev_salaries), 
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')
plt.tight_layout()
plt.legend()
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')
```
