```python
x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
# a = np.random.randint(1,10,size=20)
```

```python
plt.scatter(x,y,s=200, c='green', marker='.', edgecolor='black', linewidth =1, alpha=0.75) # s = plot size
```

```python
colors = [7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5] # you can use colormap to define this
```

```python
plt.scatter(x,y,s=200, c=colors, marker='.', edgecolor='black', linewidth =1, alpha=0.75)
```

```python
plt.scatter(x,y,s=200, c=colors, cmap='Greens', marker='.', edgecolor='black', linewidth =1, alpha=0.75)
```

```python
plt.scatter(x,y,s=200, c=colors, cmap='Greens', marker='.', edgecolor='black', linewidth =1, alpha=0.75)
cbar = plt.colorbar() # scatter 뒤에 정의해야 함
cbar.set_label('Satisfaction')
```

```python
import random
random.seed(10)
sizes = np.random.randint(150, 550, size=20)
```

```python
plt.scatter(x,y,s=sizes, c=colors, cmap='Greens', marker='.', edgecolor='black', linewidth =1, alpha=0.75)
cbar = plt.colorbar() # scatter 뒤에 정의해야 함
cbar.set_label('Satisfaction')
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Matplotlib-schafer/2019-05-31-data.csv')
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.scatter(view_count,likes, edgecolor='black', linewidth =1, alpha=0.75)
plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
```

```python
plt.scatter(view_count,likes, edgecolor='black', linewidth =1, alpha=0.75)
plt.xscale('log') # log scale로 변경
plt.yscale('log')
plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
```

```python
plt.scatter(view_count,likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth =1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label('LikeDislike Ratio')
plt.xscale('log') # log scale로 변경
plt.yscale('log')
plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
```
